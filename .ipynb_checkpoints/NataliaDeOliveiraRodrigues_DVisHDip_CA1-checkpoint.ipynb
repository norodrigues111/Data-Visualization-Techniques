# What are the top 5 “average rated” games? 

### Rationalization of the specific method to engineer the data and how it was achieved in python:

> To plot the top 5 average rated games the steps below were followed:  The dataset was sorted using sort_values(ascending=False), the function head() was used to save in a new dataframe only the TOP 5, and the average rating values were rounded to 2 decimal places, so the numbers displayed would be easier to read.
One of the top 5 games has a long name, to improve the visualization, the reduction of the characters was performed. To avoid displaying a name that could cause duality among the games, the function str.startswith() was used to check if another game has the same initial words. 


```python
top5_avg_rated_df = df.sort_values(by='average_rating', ascending=False).head()
top5_avg_rated_df['average_rating'] = top5_avg_rated_df['average_rating'].round(2)
top5_avg_rated_df = top5_avg_rated_df[['name','average_rating']]
top5_avg_rated_df
```

```python
names_check = df[df['name'].str.startswith(('The Greatest Day'))]
names_check = names_check['name'].sort_values()
names_check
```

# What are the top 5 “average rated” games? 

### Rationalization of the specific visualization to illustrate the answer:

> Once the aim is to show the top 5 average rated, the visualization will involve categorical and numerical variables.  According to (seaborn.pydata.org, 2023), Barplot() is one of the ways to visualize the relationship when one of the main variables is categorical. A bar plot also allows for easy comparison of the average ratings.    

```python
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')

ax = sns.barplot(data=top5_avg_rated_df, x='name',y='average_rating',palette='viridis')
ax.bar_label(ax.containers[0], fontsize=14,color='black',fontweight='bold')

split_labels = [label.replace(' ', '\n') for label in games_names]
ax.set_xticklabels(split_labels,fontsize=12)

ax.set_xlabel('Game Names', fontsize=14)
ax.set_ylabel('Average Rating', fontsize=14)
ax.set_title('Top 5 Average Rated Games', fontsize=16,fontweight='bold')

plt.tight_layout()
plt.show()
```

# Is there a correlation between the “users_rated” and the “max_playtime”?

### Rationalization of the specific method to engineer the data and how it was achieved in python:

> The correlation was calculated using Perason's correlation coefficient. The result close to zero implies no significant correlation.

```python
corr_users_playtime = np.corrcoef(df['users_rated'],df['max_playtime'])[0,1]
print('Correlation:',corr_users_playtime)
```

 # Is there a correlation between the “users_rated” and the “max_playtime”?
 
 ### Rationalization of the specific visualization to illustrate the answer:

> The Scatter plot is well-known as a suitable visualization to assess the relationship between two numerical variables. The choice was made because a scatter plot allows us to represent each one of the games in the dataset as a point in the visualization. In case of correlation between the variables, it’s easy to identify the pattern formed by the points. If the variables are not correlated, the points will be scattered randomly without forming a clear pattern.  

```python
plt.figure(figsize=(10,6))

plt.scatter(df['users_rated'],df['max_playtime'], alpha=0.4, edgecolors='none',s=35,color='#2e6f8e',marker='o')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Number of Users Rated',fontsize=14)
plt.ylabel('Maximum Recommended Playtime (Min)',fontsize=14)  
plt.title('Correlation Between The Users Rated and Maximum Recommended Playtime',fontsize=16,fontweight='bold') 

text = f'Correlation: {corr_users_playtime:.4f}'
plt.text(10**4, 10**4, text, fontsize=12, color='red',fontweight='bold')

x_ticks = [100, 1000, 10000,70000]  
plt.xticks(x_ticks, ['100','1,000','10,000','70,000'])  

y_ticks = [10, 100, 1000, 10000,60000]
plt.yticks(y_ticks, ['10','100','1,000','10,000','60,000'])

plt.tick_params(axis='both', which='both', width=0, labelsize=12)

ax = plt.gca()
ax.spines[['top','right']].set_visible(False)  
ax.spines[['bottom','left']].set_linewidth(1.5)  

plt.grid(True,alpha=0.3)
plt.show()
```

# What is the distribution of game categories?

### Rationalization of the specific method to engineer the data and how it was achieved in python:

> Almost 4.000 distinct categories can be found in the board_games dataset. In order to focus on the principal categories, only the 15 most frequent categories are plotted. The "80:20 rule" concept, also known as Pareto Principle is applied. According to (Anon., 2016) this rule states that 20% of the activities or causes may determine 80% of the outcomes. It is a guideline, not a rigid mathematical principle. It was achieved in Python using pd.pivot_table() to count the distribution of the game categories and saved on a new variable called category_count. The variable was sorted using sort_values(ascending=False), so the bigger numbers would be on the head of this new dataframe. The index was reset, and only the top 15 categories were saved using head(15).

```python
df.category.nunique()
```

```python
df.category.count()
```

```python
category_count = pd.pivot_table(data=df, index='category', aggfunc='size')
category_count = category_count.sort_values(ascending=False).reset_index().head(15)
category_count.columns = ['Category', 'Distribution']
category_count
```

```python
category_count.Distribution.count()
```

```python
percen_represented_80_20_rule = category_count.Distribution.sum() / df.category.count() * 100
percen_represented_80_20_rule
```

# What is the distribution of game categories?

### Rationalization of the specific visualization to illustrate the answer:

> The top 15 game categories are distributed properly in the treemap display, allowing users to instantly identify the most significant ones and their respective shares. In order to ensure that the data is both appealing and informative, an interactive visualization is provided.


```python
fig = px.treemap(category_count, path=['Category'], values='Distribution',
                 color='Distribution',  
                 color_continuous_scale=px.colors.sequential.Viridis,  
                 width=970, height=750)

fig.update_traces(textinfo='label+value+percent entry', hovertemplate=None)

fig.update_layout(title_text="<b>Distribution of Top 15 Game Categories",
                  title_font=dict(family="Arial", size=20, color="Black"),
                 title_x=0.5)

fig.show()
```

# Do older games (1992 and earlier) have a higher MEDIAN “average rating” than newer games (after 1992)?

### Rationalization of the specific visualization to illustrate the answer:

> The boxplot displays the distribution of the data in a structured and concise way. It is really useful to display the answer to the question above once the median of the data is displayed as a line inside the box. The box plot also allows us to plot side by side both dataframes representing older_games and newer_games which makes it easier to compare the median average ratings. 


```python
plt.figure(figsize=(10, 6))
plt.boxplot([older_games['average_rating'], newer_games['average_rating']], 
            labels=["Older Games (<=1992)", "Newer Games (>1992)"],
           patch_artist=True,
                     boxprops={'facecolor': '#bddf26'}, 
                     medianprops={'color': 'red','linewidth': 2})

plt.ylabel("Average Rating",fontsize=14)
plt.title("Comparison of Median Average Ratings for Older and Newer Games", fontsize=16,fontweight='bold')

ax = plt.gca()
ax.spines[['top','right','bottom','left']].set_linewidth(1.5)  

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

# “Statistically Relevant” question:

## Is there a statistically significant difference in the newer game average ratings between games where the minimum age recommended is equal or lower than 10 and games where the  minimum age recommended is equal or higher than 11, and does the year of publication (>1992) have a significant  impact on that?

### Rationalization of the specific method to engineer the data and how it was achieved in python:

> Two dataframes were created "equal_or_lower_10" and "equal_or_higher_11" slicing the dataset into games published after 1992 where the minimum age is equal or lower than 10 years old and equal or higher than 11 years old. 
After creating the new dataframes, a line plot was plotted to check if there was a difference between the average ratings by year. After plotting the line plot, a T-test was performed and the p-value of 1.0178865604825051e-152 shows that There is a statistically significance difference in the average rating between the age groups. Both dataframes were combined using pd.concat() and a pivot table was created with the mean of each group by year using pivot_table(), and saved in a new dataframe called heatmap_df.

```python
equal_or_lower_10 = df[(df['min_age']<=10) & (df['year_published']>1992)]
equal_or_higher_11 = df[(df['min_age']>10) & (df['year_published']>1992)]
```

```python
plt.figure(figsize=(10,6))

plt.plot(equal_or_lower_10.groupby('year_published')['average_rating'].mean(), label='Age 10 or lower')
plt.plot(equal_or_higher_11.groupby('year_published')['average_rating'].mean(),label='Age 11 or higher')

plt.title('Average Rating Over Time by Age Group', fontsize=16,fontweight='bold')
plt.xlabel('Year Published',fontsize=14)
plt.ylabel('Average Rating',fontsize=14)

ax = plt.gca()
ax.spines[['top','right']].set_visible(False)  
ax.spines[['bottom','left']].set_linewidth(1.5)  

plt.legend()
plt.grid(linestyle="-", alpha=0.5)
plt.show()
```

```python
t_stat, p_value = ttest_ind(equal_or_lower_10['average_rating'],equal_or_higher_11['average_rating'])

if p_value >= 0.05:
    print(f'Test Result: {p_value}. There is no statistically significance difference in the average rating between the age groups.')
else:
    print(f'Test Result: {p_value}. There is statistically significance difference in the average rating between the age groups.')
```

```python
age_group_df = pd.concat([equal_or_lower_10,equal_or_higher_11],keys=['Age 10 or lower','Age 11 or higher'],
                         names=['Age Group']).reset_index(level=0)
```

```python
heatmap_df = age_group_df.pivot_table(index='year_published',columns='Age Group',values='average_rating',aggfunc='mean')
heatmap_df.head()
```
